# Визуализация RPC curve и ROC curve (на синтетических данных)

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.metrics import precision_recall_curve, roc_curve
```

```python
# gen data
n = 10
y = np.hstack((np.zeros(n), np.ones(n)))
score_ideal = np.hstack((np.random.rand(n)*0.5, np.random.rand(n)*0.5+0.5))
score_random = np.random.rand(2*n)
score_const = np.zeros(2*n) + 0.7
score_usual = np.hstack((np.random.rand(n)*0.7, np.random.rand(n)*0.7+0.3))
```

```python
# plot data
plt.figure(figsize=(16, 4))
for i, (score, label) in enumerate(zip([score_ideal, score_random, score_const, score_usual],
                        ["ideal", "random", "const", "usual"])):
    plt.subplot(1, 4, i+1)
    plt.scatter(y, score)
    plt.title(label)
    plt.xlabel("true y")
    plt.ylabel("predicted score")
    plt.ylim(0, 1)
```

```python
plt.figure(figsize=(16, 4))
for i, (score, label) in enumerate(zip([score_ideal, score_random, score_const, score_usual],
                        ["ideal", "random", "const", "usual"])):
    plt.subplot(1, 4, i+1)
    fpr, tpr, _ = roc_curve(y, score)
    plt.plot(fpr, tpr)
    plt.title(label)
    plt.xlabel("recall")
    plt.ylabel("precision")
    plt.ylim(-0.1, 1.1)
    plt.xlim(-0.1, 1.1)
```

```python
plt.figure(figsize=(16, 4))
for i, (score, label) in enumerate(zip([score_ideal, score_random, score_const, score_usual],
                        ["ideal", "random", "const", "usual"])):
    plt.subplot(1, 4, i+1)
    precision, recall, _ = precision_recall_curve(y, score)
    plt.plot(recall, precision)
    plt.title(label)
    plt.xlabel("recall")
    plt.ylabel("precision")
    plt.ylim(-0.1, 1.1)
    plt.xlim(-0.1, 1.1)
```

From sklearn documentation: 

The last precision and recall values are 1. and 0. respectively and do not have a corresponding threshold. This ensures that the graph starts on the y axis.
